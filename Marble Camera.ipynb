```python
import hou

obj = hou.node('/obj')
geo_node = obj.createNode('geo', 'Marble_Noise_Texture')

grid_node = geo_node.createNode('grid', 'Base_Grid')
noise1_node = geo_node.createNode('attribnoise::2.0', 'add_noise1')
noise2_node = geo_node.createNode('attribnoise::2.0', 'add_noise2')
uvunwrap_node = geo_node.createNode('uvunwrap', 'create_UVs')
OUT_node = geo_node.createNode('null', 'OUT_geo')

noise1_node.setInput(0, grid_node, 0)
noise2_node.setInput(0, noise1_node, 0)
uvunwrap_node.setInput(0, noise2_node, 0)
OUT_node.setInput(0, uvunwrap_node, 0)
geo_node.layoutChildren()

OUT_node.setDisplayFlag(True)
OUT_node.setRenderFlag(True)
grid_node.setDisplayFlag(False)
grid_node.setRenderFlag(False)

grid_node.parm('rows').set(512)
grid_node.parm('cols').set(512)
noise1_node.parm('attribtype').set(0)
noise1_node.parm('amplitude').set(2.50)
noise1_node.parm('elementsize').set(1.22)
noise1_node.parm('offset').set(4.8)
noise2_node.parm('attribtype').set(0)
noise2_node.parm('basis').set(7)
noise2_node.parm('amplitude').set(7.8)
noise2_node.parm('operation').set(4)
noise2_node.parm('elementsize').set(5.33)
noise2_node.parm('offset').set(2.82)

```

```python
import hou

def create_simple_camera_rig():

#chatgpt helped, it semi-works

    camera = hou.node('/obj').createNode('cam', 'SimpleCamera')
    camera.parm('resx').set(1920)  
    camera.parm('resy').set(1080)  


    control_null = hou.node('/obj').createNode('null', 'CameraControls')
    control_null.setDisplayFlag(True)  

    camera.setInput(0, control_null)

    control_null.addSpareParmTuple(hou.FloatParmTemplate('focal_length', 'Focal Length', 1))
    control_null.addSpareParmTuple(hou.FloatParmTemplate('aperture', 'Aperture', 1))

    camera.parmTuple('t').set((0, 0, 5))
    camera.parmTuple('r').set((0, 0, 0))

    control_null.addSpareParmTuple(hou.FloatParmTemplate('translate_x', 'Translate X', 1))
    control_null.addSpareParmTuple(hou.FloatParmTemplate('translate_y', 'Translate Y', 1))
    control_null.addSpareParmTuple(hou.FloatParmTemplate('translate_z', 'Translate Z', 1,))
    control_null.addSpareParmTuple(hou.FloatParmTemplate('rotate_x', 'Rotate X', 1))
    control_null.addSpareParmTuple(hou.FloatParmTemplate('rotate_y', 'Rotate Y', 1))
    control_null.addSpareParmTuple(hou.FloatParmTemplate('rotate_z', 'Rotate Z', 1))

    control_null.parm('translate_x').setExpression('ch("tx")')
    control_null.parm('translate_y').setExpression('ch("ty")')
    control_null.parm('translate_z').setExpression('ch("tz")')
    control_null.parm('rotate_x').setExpression('ch("rx")')
    control_null.parm('rotate_y').setExpression('ch("ry")')
    control_null.parm('rotate_z').setExpression('ch("rz")')

create_simple_camera_rig()
```
